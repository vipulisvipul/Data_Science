```python
from bs4 import BeautifulSoup
from selenium import webdriver
import os
path_folder='C:/Users/Administrator/Downloads/'
os.chdir(path_folder)
```

```python
driver=webdriver.Firefox()
url1='https://en.wikipedia.org/wiki/List_of_Category_5_Atlantic_hurricanes'
driver.get(url1)
html1=driver.page_source
soup1=BeautifulSoup(html1,'html.parser')
x=soup1.find_all('table',class_='wikitable')

```

```python
import pandas as pd
```

```python
name=[]
dates=[]
duration=[]
windspeed=[]
pressure=[]
areas=[]
deaths=[]
damage=[]
for i in x[0].find_all('tr')[1:]:
    data=i.find_all('td')
    name.append(data[0].text.strip())
    dates.append(data[1].text.strip())
    duration.append(data[2].text.strip())
    windspeed.append(data[3].text.strip())
    pressure.append(data[4].text.strip())
    areas.append(data[5].text.strip())
    deaths.append(data[6].text.strip())
    damage.append(data[7].text.strip())
    
    
```

```python
y=pd.DataFrame({'name':name,'dates':dates,'duration':duration,'windspeed':windspeed,'pressure':pressure,'areas':areas,'deaths':deaths,'damage':damage})
y.head()
```

```python
y['windspeed'].str.extract(r'(\(\d{1,3})')[0].str.extract(r'(\d{1,3})').astype('int').mean()
```
