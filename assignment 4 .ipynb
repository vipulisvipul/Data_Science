### Determine your next move

```python
step=0
R=np.random.randint(1,7)
if (R==1 or R==2):
    step=step-1
elif(R==3 or R==4 or R==5):
    step=step+1
else:
    R=np.random.randint(1,7)
    step=step+R
    
print(step)
print('dice value:',R)
```

### The next step

```python
random_walk=[0]
for i in range(1,101):
    R=np.random.randint(1,7)
    if (R==1 or R==2):
        step=random_walk[-1]
        random_walk.append(step-1)
    elif(R==3 or R==4 or R==5):
        step=random_walk[-1]
        random_walk.append(step+1)
    else:
        R=np.random.randint(1,7)
        step=random_walk[-1]
        random_walk.append(R+step)
        


print(random_walk)
```

### How low can you go?

```python
random_walk=[0]
for i in range(1,101):
    R=np.random.randint(1,7)
    if (R==1 or R==2):
        step=random_walk[-1]
        random_walk.append(max(0,step-1))
    elif(R==3 or R==4 or R==5):
        step=random_walk[-1]
        random_walk.append(step+1)
    else:
        R=np.random.randint(1,7)
        step=random_walk[-1]
        random_walk.append(R+step)
        


print(random_walk)
```

### Visualize the walk

```python
import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.show()
```

### Simulate multiple walks

```python
all_walks=[]
for i in range(10):
    random_walk=[0]
    for i in range(1,101):
        R=np.random.randint(1,7)
        if (R==1 or R==2):
            step=random_walk[-1]
            random_walk.append(max(0,step-1))
        elif(R==3 or R==4 or R==5):
            step=random_walk[-1]
            random_walk.append(step+1)
        else:
            R=np.random.randint(1,7)
            step=random_walk[-1]
            random_walk.append(R+step)
    all_walks.append(random_walk)
```

### Visualize all walks

```python
np_aw=np.array(all_walks)
plt.plot(np_aw)
plt.show()
#print(np_aw)
```

```python
np_aw_t=np.transpose(np_aw)
plt.plot(np_aw_t)
plt.show()
```

### Implement clumsiness

```python
import numpy as np
np.random.seed()
all_walks=[]
for i in range(250):
    randomwalk=[0]
    for i in range(100):
        dice=np.random.randint(1,7)
        if(dice==1 or dice==2):
            step=randomwalk[-1]
            step=max(0,step-1)
            #print("The value of dice is ",dice,"was one step down",step)
        elif (dice== 3 or dice==4 or dice==5):
            step=randomwalk[-1]
            step=step+1
            #print("The value of dice is ",dice,"was one step up",step)
        else:
            step=randomwalk[-1]
            dice=np.random.randint(1,7)
            step=step+dice
            #print("The value of dice is ",dice,"and",step,"steps up")
        if np.random.rand()<0.001:
            step=0
        randomwalk.append(step)
    all_walks.append(randomwalk)
#print(all_walks)
plt.plot(np.transpose(np.array(all_walks)))
plt.show()
```
